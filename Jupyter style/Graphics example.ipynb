```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

def criar_grafico(x, y, titulo, rotulo_x, rotulo_y):
    plt.plot(x, y)
    plt.title(titulo)
    plt.xlabel(rotulo_x)
    plt.ylabel(rotulo_y)
    plt.show()

# Exemplo de dados
x_valores = [1, 2, 3, 4, 5]
y_valores = [2, 4, 6, 8, 10]

# Chamada da função para criar o gráfico
criar_grafico(x_valores, y_valores, "Gráfico Genérico", "Eixo X", "Eixo Y")
```
